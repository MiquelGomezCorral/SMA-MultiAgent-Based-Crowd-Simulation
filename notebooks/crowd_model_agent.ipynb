### Alternative: Multiprocessing-safe batch run

````python
CONFIGS =[
    Configuration(
        n_experiments=1000,
        initial_agents=250,
        width=30,
        height=30,
        agent_types_ratios = {
            "polite": 0.5,
            "aggressive": 0.3,
            "slow": 0.2
        },
        track_last_steps=4,
        path_finding_algorithm="A*",
        differentiate_exits=False,
        scenario_type="SNAKE",
        n_exits=1,
    ),
    Configuration(
        n_experiments=1000,
        initial_agents=250,
        width=30,
        height=30,
        agent_types_ratios = {
            "polite": 0.5,
            "aggressive": 0.3,
            "slow": 0.2
        },
        track_last_steps=4,
        path_finding_algorithm="BFS",
        differentiate_exits=False,
        scenario_type="SNAKE",
        n_exits=1,
    ),
    Configuration(
        n_experiments=1000,
        initial_agents=250,
        width=30,
        height=30,
        agent_types_ratios = {
            "polite": 0.5,
            "aggressive": 0.3,
            "slow": 0.2
        },
        track_last_steps=4,
        path_finding_algorithm="MANHATTAN",
        differentiate_exits=False,
        scenario_type="SNAKE",
        n_exits=1,
    ),
]
````

````python
import mesa 
import sys
import numpy as np
import pandas as pd

from maikol_utils.print_utils import print_separator, print_color

for idx, CONFIG in enumerate(CONFIGS):
    print_separator(f"Running configuration {idx + 1}/{len(CONFIGS)}", sep_type="START")
    rng = np.random.default_rng(CONFIG.seed)
    seed_values = rng.integers(0, sys.maxsize, size=(CONFIG.n_experiments,))

    # Create batch parameters from the current CONFIG
    batch_params = {
        "initial_agents": CONFIG.initial_agents,
        "width": CONFIG.width,
        "height": CONFIG.height,
        "track_last_steps": CONFIG.track_last_steps,
        "path_finding_algorithm": CONFIG.path_finding_algorithm,
        "differentiate_exits": CONFIG.differentiate_exits,
        "polite_ratio": CONFIG.agent_types_ratios["polite"],
        "aggressive_ratio": CONFIG.agent_types_ratios["aggressive"],
        "slow_ratio": CONFIG.agent_types_ratios["slow"],
        "scenario_type": CONFIG.scenario_type,
        "n_exits": CONFIG.n_exits,
    }

    results = mesa.batch_run(
        model_cls=CrowdModelWrapper,
        parameters=batch_params,
        rng=seed_values.tolist(),
        max_steps=2000,
        number_processes=16,
        data_collection_period=1,
        display_progress=True,
    )
    # len(results)
    results_df = pd.DataFrame(results)
    results_df.groupby(['RunId', 'iteration'])["Step"].max().describe()
    print_color(f"Results for configuration {idx + 1}:", color="cyan")

    print(results_df.groupby(['RunId', 'iteration'])["Step"].max().describe())
````
